```python
from collections import defaultdict

def getParams (f, count=10):
    
    linecount, wordcount = 0,0
    result = defaultdict(int)
    soubor = open(f, 'r')
    for line in soubor.readlines():
        #print("radek: {}".format(line.strip()))
        linecount += 1
        line = line.strip().split()
        wordcount += len(line)
        for word in line:
            result[word]+=1
    soubor.close()
    top = {}
    #print("Soubor {} ma {} radku a {} slov".format(file, linecount, wordcount))
    #pojmenovany atribut key
    '''
    for word, freq in sorted(result.items(), key=lambda x:x[1], reverse=True)[:count]:
        top.update({word:freq})
    '''
    top = {word:freq for word, freq in sorted(result.items(), key=lambda x:x[1], reverse=True)[:count]}
    return (linecount, wordcount, top)
```

```python
file = r'words.tst'
getParams(file, 15)
```
